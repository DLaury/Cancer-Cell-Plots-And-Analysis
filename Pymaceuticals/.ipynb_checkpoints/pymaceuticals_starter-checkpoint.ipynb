```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mDrugDF = pd.read_csv(mouse_drug_data_to_load)
clinicalTrialsDF = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combinedMiceDF = pd.merge(clinicalTrialsDF, mDrugDF, on="Mouse ID")

# Display the data table for preview
combinedMiceDF.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
tumorVolData = combinedMiceDF.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
tumorDF = pd.DataFrame(tumorVolData)

# Preview DataFrame
tumorDF.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
stdError = combinedMiceDF.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
errorDF = pd.DataFrame(stdError)

# Preview DataFrame
errorDF.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
unstackedTumorDF = tumorDF.unstack(0)
unstackedErrorDF = errorDF.unstack(0)

# Preview that Reformatting worked
unstackedTumorDF.head()
```

```python
unstackedTumorDF["Tumor Volume (mm3)"]["Capomulin"].plot(kind='line', marker='o', linestyle="--",
                    yerr=unstackedErrorDF["Tumor Volume (mm3)"]["Capomulin"], title="Tumor Response Treatment")
unstackedTumorDF["Tumor Volume (mm3)"]["Infubinol"].plot(kind='line', marker='^', linestyle="--",
                    yerr=unstackedErrorDF["Tumor Volume (mm3)"]["Infubinol"])
unstackedTumorDF["Tumor Volume (mm3)"]["Ketapril"].plot(kind='line', marker='s', linestyle="--",
                    yerr=unstackedErrorDF["Tumor Volume (mm3)"]["Ketapril"])
unstackedTumorDF["Tumor Volume (mm3)"]["Placebo"].plot(kind='line', marker='d', grid=True, linestyle="--",
                    yerr=unstackedErrorDF["Tumor Volume (mm3)"]["Placebo"])
plt.legend()
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint
metaSites = combinedMiceDF.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame
metaSitesDF = pd.DataFrame(metaSites)

# Preview DataFrame
metaSitesDF.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint
metaError = combinedMiceDF.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame
metaErrorDF = pd.DataFrame(metaError)

# Preview DataFrame
metaErrorDF.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
unstackedMetaSitesDF = metaSitesDF.unstack(0)
unstackedMetaErrorDF = metaErrorDF.unstack(0)

# Preview that Reformatting worked
unstackedMetaSitesDF.head()
```

```python
# Generate the Plot (with Error Bars)
unstackedMetaSitesDF["Metastatic Sites"]["Capomulin"].plot(kind='line', marker='o', linestyle="--",
                    yerr=unstackedMetaErrorDF["Metastatic Sites"]["Capomulin"], title="Tumor Response Treatment")
unstackedMetaSitesDF["Metastatic Sites"]["Infubinol"].plot(kind='line', marker='^', linestyle="--",
                    yerr=unstackedMetaErrorDF["Metastatic Sites"]["Infubinol"])
unstackedMetaSitesDF["Metastatic Sites"]["Ketapril"].plot(kind='line', marker='s', linestyle="--",
                    yerr=unstackedMetaErrorDF["Metastatic Sites"]["Ketapril"])
unstackedMetaSitesDF["Metastatic Sites"]["Placebo"].plot(kind='line', marker='d', grid=True, linestyle="--",
                    yerr=unstackedMetaErrorDF["Metastatic Sites"]["Placebo"])
plt.legend()
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
miceCount = combinedMiceDF.groupby(["Drug", "Timepoint"])["Mouse ID"].count()

# Convert to DataFrame
miceCountDF = pd.DataFrame(miceCount)

# Preview DataFrame
miceCountDF.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
unstackedMiceCountDF = miceCountDF.unstack(0)

# Preview the Data Frame
unstackedMiceCountDF.head()
```

```python
# Generate the Plot (Accounting for percentages)
startingCount = unstackedMiceCountDF.iloc[0]
percentageMiceDF=(100 * unstackedMiceCountDF / startingCount).round(2)
percentageMiceDF["Mouse ID"]["Capomulin"].plot(kind='line', marker='o', linestyle="--",
                     title="Tumor Response Treatment")
percentageMiceDF["Mouse ID"]["Infubinol"].plot(kind='line', marker='^', linestyle="--")
percentageMiceDF["Mouse ID"]["Ketapril"].plot(kind='line', marker='s', linestyle="--")
percentageMiceDF["Mouse ID"]["Placebo"].plot(kind='line', marker='d', grid=True, linestyle="--")
plt.legend()
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
lengthTumorDF = len(unstackedTumorDF)
percentTumorDF = unstackedTumorDF.pct_change(periods=(lengthTumorDF-1))
percentages = percentTumorDF.tail(1).stack().reset_index()
percentages["Tumor Volume (mm3)"] = (percentages["Tumor Volume (mm3)"] * 100).round(2)
percentages
```

```python
percents = [percentages["Tumor Volume (mm3)"][0], percentages["Tumor Volume (mm3)"][2],
           percentages["Tumor Volume (mm3)"][3], percentages["Tumor Volume (mm3)"][5]]
xaxis = np.arange(len(percents))
barLabels = [percentages["Drug"][0], percentages["Drug"][2],
           percentages["Drug"][3], percentages["Drug"][5]]
plt.grid(alpha=.5)
barPlot = plt.bar(x=xaxis, height=percents, align='center', tick_label=barLabels, zorder=2)
plt.axhline(0, color='k', zorder=3)
barPlacement=-.215
i = 0

for value in percents:
    if value > 1:
        plt.text(barPlot[i].get_x() + barPlot[i].get_width() / 2 ,2 , str(percents[i]) + "%", 
                 name = 'Helvetica', size = 'large', color = 'w', ha = 'center')
        barPlot[i].set_color('lightcoral')
        i = i + 1
    else:
        plt.text(barPlot[i].get_x() + barPlot[i].get_width() / 2 ,-5, str(percents[i]) + "%",
                 name = 'Helvetica', size='large', ha = 'center')
        barPlot[i].set_color('lightgreen')
        i = i + 1
plt.ylim(-20,60)
plt.xlabel("Drug")
plt.ylabel("Tumor Volume Change (%)")
plt.title("Tumor Change Over Treatment (45 Days)", weight = 'bold')
```
